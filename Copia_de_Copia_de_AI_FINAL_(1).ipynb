#### Creacion de la funcion basada en una interaccion conversacional con el modelo.

```python
import re
import json # Import the json module
def buscar_datasets_con_ia(consulta):
    prompt = f"""
    Encuentra datasets de datos abierotos y públicos relacionados con esta búsqueda: '{consulta}'.
    Los datasetes deben ser descargables en este repositorio '{repositorio}'.
    Devuélveme solo una lista JSON con máximo 10 resultados con esta estructura:
    [
      {{
        "ID": 1,
        "repositorio": "Kaggle",
        "titulo": "Nombre del dataset",
        "autor": "Autor",
        "año": "2023"
      }},
      ...
    ]
    """ # Se construye la pregunta especifica 'el prompt' con el que se va a interactuar con el modelo

    respuesta = modelo.generate_content(prompt) # Se define la interaccion con el modelo
    texto = respuesta.text # Extraccion del texto con la respuesta Gemini
    json_match = re.search(r"\[.*\]", texto, re.DOTALL) # El resto logica de extraccion y procesamiento JSON
    if json_match:
        try:
            return json.loads(json_match.group(0))
        except json.JSONDecodeError:
            print("⚠️ Error al decodificar JSON:")
            print(texto)
            return []
    else:
        print("⚠️ No se encontró JSON:")
        print(texto)
        return []

# --- Ejecutar búsqueda ---
consulta = input("🔍 ¿Qué tipo de dataset estás buscando?: ") # Especificar la primera condicion de la consulta
repositorio = input("🔍 ¿En que repositorio estás buscando?: ") # Especificar la segunda condicion de la consulta dos
datasets = buscar_datasets_con_ia(consulta) # Almacenamiento de los resultados con la interaccion de Gemini

if not datasets:
    print("❌ No se encontraron resultados.")
else:
    print("\n✅ Resultados encontrados:")
    for ds in datasets:
        print(f"[{ds['ID']}] {ds['titulo']} ({ds['repositorio']}, {ds['año']}) - Autor: {ds['autor']}")

    # Guardar en archivo para usar luego
    with open("resultados_datasets.json", "w", encoding="utf-8") as f:
        json.dump(datasets, f, indent=2, ensure_ascii=False)
    print("\n📁 Resultados guardados en 'resultados_datasets.json'")
```

#### Creacion de la funcion para la consulta de la meta data

```python
def analizar_metadata_con_ia(autor, repositorio, titulo, año):
    prompt = f"""
    Eres un asistente experto en datasets públicos.
    Basado en el dataset titulado '{titulo}', del repositorio '{repositorio}', creado por '{autor}' en el año {año},
    realiza un análisis estimado de la metadata interna del dataset y una breve descripción general (resumen) del contenido.

    Devuelve un JSON con esta estructura:
    {{
      "titulo": "{titulo}",
      "repositorio": "{repositorio}",
      "autor": "{autor}",
      "año": "{año}",
      "resumen": "breve resumen descriptivo del dataset",
      "numero_registros": número entero estimado,
      "numero_variables": número entero estimado,
      "nombres_variables": [lista de strings con nombres de variables],
      "tipos_variables": [lista de strings con tipos de variables (ej: numérico, texto, fecha)],
      "comentarios": "observaciones adicionales relevantes del dataset"
    }}

    SOLO responde con el JSON.
    """ # Otra vez pedirle a Gemini la estimacion de la metadata de un data set, agregando algunos itmes claves, entre ellos, la asignacion de un identificador.
    respuesta = modelo_metadata.generate_content(prompt) # Se define la interaccion con el modelo
    texto = respuesta.text # Extraccion del texto con la respuesta Gemini
    json_match = re.search(r"\{.*\}", texto, re.DOTALL)# El resto logica de extraccion y procesamiento JSON
    if json_match:
        try:
            return json.loads(json_match.group(0))
        except json.JSONDecodeError:
            print("⚠️ Error al decodificar JSON:")
            print(texto)
            return None
    else:
        print("⚠️ No se encontró JSON en la respuesta:")
        print(texto)
        return None

# --- Cargar resultados previos ---
try:
    with open("resultados_datasets.json", "r", encoding="utf-8") as f:
        datasets = json.load(f)
except FileNotFoundError:
    print("❌ No se encontró 'resultados_datasets.json'. Ejecuta primero el flujo de búsqueda.")
    exit()

# --- Selección por ID ---
id_elegido = int(input("\n🔢 Ingresa el ID del dataset para analizar su metadata: "))
seleccionado = next((d for d in datasets if d["ID"] == id_elegido), None)

if seleccionado:
    print("\n🔍 Consultando metadata estimada para:")
    print(f"📌 Título: {seleccionado['titulo']}")
    print(f"📁 Repositorio: {seleccionado['repositorio']}")
    print(f"👤 Autor: {seleccionado['autor']}")
    print(f"📅 Año: {seleccionado['año']}")

    metadata = analizar_metadata_con_ia(
        autor=seleccionado["autor"],
        repositorio=seleccionado["repositorio"],
        titulo=seleccionado["titulo"],
        año=seleccionado["año"]
    )

    if metadata:
        print("\n📊 Metadata estimada:")
        print(json.dumps(metadata, indent=2, ensure_ascii=False))
    else:
        print("❌ No se pudo analizar la metadata.")
else:
    print("❌ ID no encontrado.")
```

```python
from duckduckgo_search import DDGS

if datasets:
    try:
        # Use the previously defined variable id_elegido instead of the undefined id_consulta
        dataset_seleccionado = next((item for item in datasets if item.get("ID") == id_elegido), None)

        if dataset_seleccionado:

            consulta_busqueda = f"site:{dataset_seleccionado.get('repositorio').lower()}.com {dataset_seleccionado.get('titulo')} dataset"

            print(f"\nBuscando enlaces reales con DuckDuckGo para: {consulta_busqueda}")

            with DDGS() as ddgs:
                resultados_ddg = list(ddgs.text(consulta_busqueda, max_results=10))

            if resultados_ddg:
                print("\nEnlaces encontrados:")
                for i, resultado in enumerate(resultados_ddg):
                    print(f"[{i}] {resultado['href']}")

                # Solo pedir que selecciones el índice del enlace que quieres usar
                id_enlace = int(input("\nSelecciona el enlace (por número) donde quieres descargar los datos: "))
                if 0 <= id_enlace < len(resultados_ddg):
                    print(f"\nSeleccionaste el enlace: {resultados_ddg[id_enlace]['href']}")
                else:
                    print("Número de enlace inválido.")
            else:
                print("No se encontraron enlaces con DuckDuckGo.")

        else:
            print("No se encontró ningún dataset con ese ID.")
    except ValueError:

        print("Por favor, ingresa un número entero válido.")
else:
    print("No hay datasets cargados.")
```

```python
# 📚 2. Importa librerías
import pandas as pd
import os
from langchain_groq import ChatGroq
# Update the import path for create_pandas_dataframe_agent
from langchain_experimental.agents import create_pandas_dataframe_agent
from langchain.prompts import PromptTemplate

# 🔐 3. Configura tu clave API de Groq
os.environ["GROQ_API_KEY"] = "GROQ_API_KEY"  # <-- Reemplaza por tu clave real

# 📂 4. Carga tu dataset CSV
# The correct path seems to be the one extracted earlier
df = pd.read_csv("/content/tweets.csv")  # <-- Use the correct file path

# 🤖 5. Carga el modelo LLaMA 3 desde Groq
llm = ChatGroq(
    groq_api_key=os.environ["GROQ_API_KEY"],
    model_name="llama3-70b-8192"
)

# 🧠 6. Crea un prompt personalizado para análisis de datos
prompt = PromptTemplate.from_template("""
Eres un experto analista de datos que responde preguntas sobre un DataFrame `df` usando pandas.

Tu objetivo es:
- Responder en español.
- No inventar datos ni columnas.
- Usar pandas internamente para los cálculos.
- Decir si una columna o dato no existe.
- Realizar análisis exploratorio simple (EDA).

Ejemplos:
Pregunta: ¿Cuántas filas hay?
Respuesta: El DataFrame tiene 1432 filas.

Pregunta: ¿Qué columnas tienen valores nulos?
Respuesta: Las columnas 'edad' y 'salario' tienen valores nulos.

Pregunta: ¿Cuáles son las columnas categóricas?
Respuesta: Son aquellas con tipo 'object' o que tienen pocos valores únicos, como 'ciudad' y 'sexo'.

Ahora responde la siguiente pregunta:
{input}
""")

# 🧠 7. Crea el agente de análisis exploratorio con pandas
agent = create_pandas_dataframe_agent(
    llm=llm,
    df=df,
    prompt=prompt,
    verbose=True,
    allow_dangerous_code=True,
    handle_parsing_errors=True # <-- Add this argument
)
```

- Configuracion de las solicitudes especificas del agente
- Impresion de los resultados

```python
# 🧾 Lista para guardar resultados
resultados = []

# ❓ 8. Ejecuta preguntas y guarda respuestas
preguntas = [
    "¿Cuántas filas y columnas tiene el dataset?",
    "¿Cuáles son los nombres de las columnas y su tipo estructura?",
    "¿Qué columnas tienen valores nulos?",
    "¿Cuál es el porcentaje de valores nulos por columna?",
    "¿Hay valores duplicados en el dataset?",
    "¿Cuáles son los valores únicos de las columnas categoricas",
    "¿Cual es el porcentaje de los valores de las columnas bool?",
    "¿Cuál es el valor máximo y mínimo de la columnas numericas ?",
    "¿Cuál es la media, mediana y desviación estándar de cada columna numérica ?"
]

for pregunta in preguntas:
    print("🔍 Pregunta:", pregunta)
    try:
        respuesta = agent.run(pregunta)
        print("🧠 Respuesta:", respuesta)
        print("=" * 100)

        # Guarda en lista
        resultados.append({
            "pregunta": pregunta,
            "respuesta": respuesta,
            "modelo": llm.model_name,# cambiar segun el modelo usado
            "archivo_analizado": "president_county_candidate.csv"# cambiar
        })
    except Exception as e: # Catching a general Exception for robustness
        print(f"❌ Error processing question '{pregunta}': {e}")
        print("=" * 100)
        # Optionally, store the error details instead of a successful response
        resultados.append({
            "pregunta": pregunta,
            "respuesta": f"Error: {e}",
            "modelo": llm.model_name,
            "archivo_analizado": "president_county_candidate.csv"
        })


# 🗃️ Convertir a DataFrame
df_resultados = pd.DataFrame(resultados)

# 💾 Guardar como CSV
df_resultados.to_csv("/content/respuestas_agente_eda.csv", index=False)
```

## **Modelado y extraccion de temas**


En este punto, luego de la exploracion de los datos, y confirmar la validez estadissticas de los mismos, y sus especificaciones, se hace posible la ejecucion de nuestro procesamiento de datos. Para ello, es necesario la creacion de un cliente que permita gestionar la conexión y la interacción con el servicio de IA. El servicio requerido es un patrón común y necesario en el desarrollo con APIs de LLMs. Es como tener un control remoto (el objeto cliente) para operar un dispositivo (el modelo de IA en la nube).

En este caso, configuramos el modelo **`"claude-3-haiku-20240307"`** de **`"Antropic"`**. Una vez configurado el modelo, crea el objeto especial llamado cliente. Este cliente servira como interfaz directa para interactuar con los servidores de Anthropic. Este objeto ya tiene toda la lógica para enviar tus solicitudes y recibir las respuestas del modelo.

Para ello, configuramos la función *analyze_text_con_claude*, un componente crucial para el procesamiento inicial y la extracción de información clave de tus textos. Esta función combina la precisión de las expresiones regulares con la capacidad de comprensión avanzada de un Modelo de Lenguaje Grande (LLM) para automatizar el análisis de contenido.

>Analiza texto para extraer hashtags, URLs, y los 3 temas principales, usando Anthropic Claude para temas y regex para lo demás.

>Args:
>   content (str): El texto que quieres analizar.

> Returns:

> dict: Un diccionario con "hashtags", "urls", y "temas".

> Incluye mensajes de error si hay problemas con la API, el formato,
> o si el contenido de entrada es inválido.

Luego, la función *analyze_text_con_claude* está diseñada para la extraccion de entidades específicas, utilizando expresiones regulares (re) con lo cual se identifico y extrajo patrones simples pero importantes (Hashtags y URLs).

La tarea más compleja, la extracción de los 3 temas más relevantes del texto, se delega al modelo Anthropic Claude (específicamente claude-3-haiku-20240307).

Instrucciones al LLM (el Prompt): La función construye un prompt detallado que se envía directamente a Claude. El prompt construido fue el siguiente:

> You are an expert text analysis assistant.
  Analyze the following text and return the 3 most relevant main themes or keywords.
  The result must be a valid JSON object with a "themes" key containing a list of strings.
  Ensure the output is *only* the JSON object, with no additional text before or after.

    JSON output example:
    {{
      "themes": ["keyword 1", "keyword 2", "keyword 3"]
      }}
    Text to analyze:"""

Este prompt le indica a Claude que actúe como un "asistente experto en análisis de texto", que analice el contenido provisto y que devuelva los 3 temas principales en un formato JSON estandarizado (una lista bajo la clave "themes").

Con este prompt, pudimos configurar la interacción con la API. Utilizamos el objeto client (previamente configurado con tu API Key de Anthropic) para enviar el prompt al modelo Claude en la nube y recibir su respuesta.

Esta conexion optimmiza el procesamiento de la Respuesta. Una vez que Claude devuelve su respuesta (esperando que sea un JSON válido), la función la parsea para extraer los temas. Se incluye un robusto manejo de errores para lidiar con textos de entrada inválidos, problemas de comunicación con la API de Claude o si Claude no devuelve el JSON en el formato esperado, se asegura que la función no falle inesperadamente.

Finalmente, la función consolida toda la información extraída (hashtags, URLs y temas) en un único diccionario, facilitando su uso en análisis posteriores.

Es importante destacar que el prompt utilizado dentro de esta función es específico para instruir a Claude sobre la tarea de extracción de temas. Esto es distinto de la plantilla de instrucciones general que se utiliza para configurar el agente de LangChain que opera sobre DataFrames. Cada plantilla cumple un rol particular para el "actor" de IA al que va dirigida. En otras palabras, definimos una plantilla de instrucciones (el prompt_instructions) dentro de tu función analyze_text_con_claude que está específicamente diseñada para guiar a Claude en la tarea de extracción de los tres temas principales y asegurar que la respuesta se registre en la estructura de diccionario (JSON) que esperas.

```python
## Configuracion del agente

import os
import re
import json
import pandas as pd
from anthropic import Anthropic


# --- 1.1. Configure Your Anthropic API Key ---

os.environ["ANTHROPIC_API_KEY"] = "ANTROPIC_API_KEY"
ANTHROPIC_API_KEY = os.environ.get("ANTHROPIC_API_KEY")


if not ANTHROPIC_API_KEY:
    raise ValueError(
        "Error! The 'ANTHROPIC_API_KEY' environment variable isn't set. "
        "Please configure it before running the code."
    )

# --- 1.2. Initialize the Anthropic Client ---
client = Anthropic(api_key=ANTHROPIC_API_KEY)

# --- 1.3. Select the Claude Model ---
ANTHROPIC_MODEL = "claude-3-haiku-20240307"

print(f"Setup complete. Using Anthropic model: {ANTHROPIC_MODEL}")
```

### Creacion de la funcion basada en una interaccion conversacional con el modelo.

```python
def analyze_text_con_claude(content: str) -> dict:
    """
    Analiza texto para extraer hashtags, URLs, y los 3 temas principales,
    usando Anthropic Claude para temas y regex para lo demás.

    Args:
        content (str): El texto que quieres analizar.

    Returns:
        dict: Un diccionario con "hashtags", "urls", y "temas".
              Incluye mensajes de error si hay problemas con la API, el formato,
              o si el contenido de entrada es inválido.
    """
    # --- ¡NUEVO! 2.1. Validación del Contenido de Entrada ---
    # Comprobamos si el contenido no es una cadena o si está vacío/solo espacios.
    if not isinstance(content, str) or not content.strip():
        # Devolvemos un diccionario de error explícito en lugar de dejar que la función falle.
        return {
            "hashtags": [],
            "urls": [],
            "temas": [f"Error: Contenido de texto inválido o vacío (tipo: {type(content).__name__})."]
        }

    # --- 2.2. Extraer Hashtags (con expresiones regulares) ---
    hashtags = list(set(re.findall(r'#(\w+)', content)))

    # --- 2.3. Extraer URLs (con expresiones regulares) ---
    url_pattern = re.compile(
        r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+'
    )
    urls = url_pattern.findall(content)

    # --- 2.4. Identificar Temas Principales (con Anthropic Claude) ---
    extracted_themes = []

    # Preparamos el "prompt" o instrucción para Claude.
    prompt_instructions = f"""
    You are an expert text analysis assistant.
    Analyze the following text and return the 3 most relevant main themes or keywords.
    The result must be a valid JSON object with a "themes" key containing a list of strings.
    Ensure the output is *only* the JSON object, with no additional text before or after.

    JSON output example:
    {{
      "themes": ["keyword 1", "keyword 2", "keyword 3"]
    }}

    Text to analyze:
    {content}
    """

    # Preparamos una versión truncada del contenido para los mensajes de error,
    # para que no imprimamos textos muy largos en la consola.
    truncated_content_for_error = content[:75] + "..." if len(content) > 75 else content


    try:
        api_response = client.messages.create(
            model=ANTHROPIC_MODEL,
            max_tokens=200,
            messages=[
                {"role": "user", "content": prompt_instructions}
            ]
        )

        response_text = api_response.content[0].text.strip()
        json_response = json.loads(response_text)

        if "themes" in json_response and isinstance(json_response["themes"], list):
            extracted_themes = json_response["themes"][:3]
        else:
            print(f"Warning: Unexpected theme format for text: {truncated_content_for_error}. Raw response: {response_text}")
            extracted_themes = [response_text]

    except json.JSONDecodeError as e:
        print(f"Warning: Claude didn't return valid JSON for text: {truncated_content_for_error}. Error: {e}")
        print(f"Raw Claude response: '{response_text}'")
        extracted_themes = ["Error: Invalid JSON from Claude.", response_text]
    except Exception as e:
        print(f"Warning: General error calling Claude API for text: {truncated_content_for_error}. Error: {e}")
        extracted_themes = ["Error extracting themes with Claude."]

    # --- 2.5. Devolver todos los resultados ---
    return {
        "hashtags": hashtags,
        "urls": urls,
        "temas": extracted_themes
    }
```

### Aplicacion del modelo en el data set muestreado.

Debido a la magnitud de registo de nuestro data frame, hemos decidido hacer un muestreo de la informacion del 10%. La muestra del 10% de los registros permitiran computar y procesar la informacion adecuada para nuestros objetivos de estudio.

De manera que, la aplicacion del analisis de texto orientado por nuestras solicitudes extraera los temas, hashtags y url's, y posteriormente, identificara los temas principales de cada registro. El resultado sera una columna adicional con la lista de url's, temas y hashtags asociado a cada texto publicado.

```python
# --- 3.2. Configuración del Muestreo (Opcional, pero recomendado para pruebas) ---
# Define cuántas filas quieres muestrear para probar.
# Si lo pones a 0, procesará todo el DataFrame.
NUM_ROWS_TO_SAMPLE = int(len(df_csv) * 0.1) # <--- ¡CAMBIA ESTE NÚMERO SEGÚN TUS NECESIDADES DE PRUEBA!

# --- 3.3. Realiza el Muestreo del DataFrame (si se configuró) ---
df_to_process = None
if 'content' not in df_csv.columns:
    print("\n¡Error! El DataFrame 'df_csv' no tiene la columna 'content'. Asegúrate de que tus textos estén en esa columna.")
elif NUM_ROWS_TO_SAMPLE > 0 and NUM_ROWS_TO_SAMPLE <= len(df_csv):
    df_to_process = df_csv.sample(n=NUM_ROWS_TO_SAMPLE, random_state=42)
    print(f"\n--- 3.3. Se ha creado una muestra de {len(df_to_process)} filas de 'df_csv' para el análisis ---")
    print(df_to_process[['content']])
else:
    print("\n--- 3.3. No se ha configurado un muestreo válido o el número excede el tamaño del DF. Se procesará 'df_csv' completo ---")
    df_to_process = df_csv

# --- 3.4. Aplicar la Función de Análisis a la MUESTRA (o al DF completo) ---
if df_to_process is not None and not df_to_process.empty:
    print(f"\n--- 3.4. Iniciando el análisis de texto (usando {ANTHROPIC_MODEL}). Esto puede tardar y generar costos de API ---")

    # Aplicamos la función 'analyze_text_con_claude' a cada texto de la columna 'content'.
    df_to_process['analysis_results'] = df_to_process['content'].apply(analyze_text_con_claude)

    print("\n--- 3.4. Análisis completado en la muestra (o DataFrame completo) ---")
    print("Primeras filas con los resultados del análisis:")
    print(df_to_process[['content', 'analysis_results']].to_string())

    print("\n--- Detalles del análisis de los primeros 5 textos analizados (ejemplo): ---")
    if not df_to_process.empty:
        for i in range(min(5, len(df_to_process))):
            print(f"\nResultado para la fila {df_to_process.index[i]}:")
            print(df_to_process['analysis_results'].iloc[i])

else:
    print("\n¡Error! El DataFrame no pudo ser preparado para el análisis. Revisa los mensajes anteriores.")
```
