#### Creacion de la funcion basada en una interaccion conversacional con el modelo.

```python
import re
import json # Import the json module
def buscar_datasets_con_ia(consulta):
    prompt = f"""
    Encuentra datasets de datos abierotos y p√∫blicos relacionados con esta b√∫squeda: '{consulta}'.
    Los datasetes deben ser descargables en este repositorio '{repositorio}'.
    Devu√©lveme solo una lista JSON con m√°ximo 10 resultados con esta estructura:
    [
      {{
        "ID": 1,
        "repositorio": "Kaggle",
        "titulo": "Nombre del dataset",
        "autor": "Autor",
        "a√±o": "2023"
      }},
      ...
    ]
    """ # Se construye la pregunta especifica 'el prompt' con el que se va a interactuar con el modelo

    respuesta = modelo.generate_content(prompt) # Se define la interaccion con el modelo
    texto = respuesta.text # Extraccion del texto con la respuesta Gemini
    json_match = re.search(r"\[.*\]", texto, re.DOTALL) # El resto logica de extraccion y procesamiento JSON
    if json_match:
        try:
            return json.loads(json_match.group(0))
        except json.JSONDecodeError:
            print("‚ö†Ô∏è Error al decodificar JSON:")
            print(texto)
            return []
    else:
        print("‚ö†Ô∏è No se encontr√≥ JSON:")
        print(texto)
        return []

# --- Ejecutar b√∫squeda ---
consulta = input("üîç ¬øQu√© tipo de dataset est√°s buscando?: ") # Especificar la primera condicion de la consulta
repositorio = input("üîç ¬øEn que repositorio est√°s buscando?: ") # Especificar la segunda condicion de la consulta dos
datasets = buscar_datasets_con_ia(consulta) # Almacenamiento de los resultados con la interaccion de Gemini

if not datasets:
    print("‚ùå No se encontraron resultados.")
else:
    print("\n‚úÖ Resultados encontrados:")
    for ds in datasets:
        print(f"[{ds['ID']}] {ds['titulo']} ({ds['repositorio']}, {ds['a√±o']}) - Autor: {ds['autor']}")

    # Guardar en archivo para usar luego
    with open("resultados_datasets.json", "w", encoding="utf-8") as f:
        json.dump(datasets, f, indent=2, ensure_ascii=False)
    print("\nüìÅ Resultados guardados en 'resultados_datasets.json'")
```

```python
def analizar_metadata_con_ia(autor, repositorio, titulo, a√±o):
    prompt = f"""
    Eres un asistente experto en datasets p√∫blicos.
    Basado en el dataset titulado '{titulo}', del repositorio '{repositorio}', creado por '{autor}' en el a√±o {a√±o},
    realiza un an√°lisis estimado de la metadata interna del dataset y una breve descripci√≥n general (resumen) del contenido.

    Devuelve un JSON con esta estructura:
    {{
      "titulo": "{titulo}",
      "repositorio": "{repositorio}",
      "autor": "{autor}",
      "a√±o": "{a√±o}",
      "resumen": "breve resumen descriptivo del dataset",
      "numero_registros": n√∫mero entero estimado,
      "numero_variables": n√∫mero entero estimado,
      "nombres_variables": [lista de strings con nombres de variables],
      "tipos_variables": [lista de strings con tipos de variables (ej: num√©rico, texto, fecha)],
      "comentarios": "observaciones adicionales relevantes del dataset"
    }}

    SOLO responde con el JSON.
    """ # Otra vez pedirle a Gemini la estimacion de la metadata de un data set, agregando algunos itmes claves, entre ellos, la asignacion de un identificador.
    respuesta = modelo_metadata.generate_content(prompt) # Se define la interaccion con el modelo
    texto = respuesta.text # Extraccion del texto con la respuesta Gemini
    json_match = re.search(r"\{.*\}", texto, re.DOTALL)# El resto logica de extraccion y procesamiento JSON
    if json_match:
        try:
            return json.loads(json_match.group(0))
        except json.JSONDecodeError:
            print("‚ö†Ô∏è Error al decodificar JSON:")
            print(texto)
            return None
    else:
        print("‚ö†Ô∏è No se encontr√≥ JSON en la respuesta:")
        print(texto)
        return None

# --- Cargar resultados previos ---
try:
    with open("resultados_datasets.json", "r", encoding="utf-8") as f:
        datasets = json.load(f)
except FileNotFoundError:
    print("‚ùå No se encontr√≥ 'resultados_datasets.json'. Ejecuta primero el flujo de b√∫squeda.")
    exit()

# --- Selecci√≥n por ID ---
id_elegido = int(input("\nüî¢ Ingresa el ID del dataset para analizar su metadata: "))
seleccionado = next((d for d in datasets if d["ID"] == id_elegido), None)

if seleccionado:
    print("\nüîç Consultando metadata estimada para:")
    print(f"üìå T√≠tulo: {seleccionado['titulo']}")
    print(f"üìÅ Repositorio: {seleccionado['repositorio']}")
    print(f"üë§ Autor: {seleccionado['autor']}")
    print(f"üìÖ A√±o: {seleccionado['a√±o']}")

    metadata = analizar_metadata_con_ia(
        autor=seleccionado["autor"],
        repositorio=seleccionado["repositorio"],
        titulo=seleccionado["titulo"],
        a√±o=seleccionado["a√±o"]
    )

    if metadata:
        print("\nüìä Metadata estimada:")
        print(json.dumps(metadata, indent=2, ensure_ascii=False))
    else:
        print("‚ùå No se pudo analizar la metadata.")
else:
    print("‚ùå ID no encontrado.")
```

## **Modelado y extraccion de temas**


En este punto, luego de la exploracion de los datos, y confirmar la validez estadissticas de los mismos, y sus especificaciones, se hace posible la ejecucion de nuestro procesamiento de datos. Esta funci√≥n combina la precisi√≥n de las expresiones regulares con la capacidad de comprensi√≥n avanzada de un Modelo de Lenguaje Grande (LLM) para automatizar el an√°lisis de contenido.

Luego, la funci√≥n *analyze_text_con_claude* est√° dise√±ada para la extraccion de entidades espec√≠ficas, utilizando expresiones regulares (re) con lo cual se identifico y extrajo patrones simples pero importantes (Hashtags y URLs).

Instrucciones al LLM (el Prompt): La funci√≥n construye un prompt detallado que se env√≠a directamente a Claude. El prompt construido fue el siguiente:

> You are an expert text analysis assistant.

Este prompt le indica a Claude que act√∫e como un "asistente experto en an√°lisis de texto", que analice el contenido provisto y que devuelva los 3 temas principales en un formato JSON estandarizado (una lista bajo la clave "themes").

Con este prompt, pudimos configurar la interacci√≥n con la API. Utilizamos el objeto client (previamente configurado con tu API Key de Anthropic) para enviar el prompt al modelo Claude en la nube y recibir su respuesta.

```python
## Configuracion del agente

import os
import re
import json
import pandas as pd
from anthropic import Anthropic


# --- 1.1. Configure Your Anthropic API Key ---

os.environ["ANTHROPIC_API_KEY"] = "ANTROPIC_API_KEY"
ANTHROPIC_API_KEY = os.environ.get("ANTHROPIC_API_KEY")


if not ANTHROPIC_API_KEY:
    raise ValueError(
        "Error! The 'ANTHROPIC_API_KEY' environment variable isn't set. "
        "Please configure it before running the code."
    )

# --- 1.2. Initialize the Anthropic Client ---
client = Anthropic(api_key=ANTHROPIC_API_KEY)

# --- 1.3. Select the Claude Model ---
ANTHROPIC_MODEL = "claude-3-haiku-20240307"

print(f"Setup complete. Using Anthropic model: {ANTHROPIC_MODEL}")
```

### Creacion de la funcion basada en una interaccion conversacional con el modelo.

```python
def analyze_text_con_claude(content: str) -> dict:
    """
    Analiza texto para extraer hashtags, URLs, y los 3 temas principales,
    usando Anthropic Claude para temas y regex para lo dem√°s.

    Args:
        content (str): El texto que quieres analizar.

    Returns:
        dict: Un diccionario con "hashtags", "urls", y "temas".
              Incluye mensajes de error si hay problemas con la API, el formato,
              o si el contenido de entrada es inv√°lido.
    """
    # --- ¬°NUEVO! 2.1. Validaci√≥n del Contenido de Entrada ---
    # Comprobamos si el contenido no es una cadena o si est√° vac√≠o/solo espacios.
    if not isinstance(content, str) or not content.strip():
        # Devolvemos un diccionario de error expl√≠cito en lugar de dejar que la funci√≥n falle.
        return {
            "hashtags": [],
            "urls": [],
            "temas": [f"Error: Contenido de texto inv√°lido o vac√≠o (tipo: {type(content).__name__})."]
        }

    # --- 2.2. Extraer Hashtags (con expresiones regulares) ---
    hashtags = list(set(re.findall(r'#(\w+)', content)))

    # --- 2.3. Extraer URLs (con expresiones regulares) ---
    url_pattern = re.compile(
        r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+'
    )
    urls = url_pattern.findall(content)

    # --- 2.4. Identificar Temas Principales (con Anthropic Claude) ---
    extracted_themes = []

    # Preparamos el "prompt" o instrucci√≥n para Claude.
    prompt_instructions = f"""
    You are an expert text analysis assistant.
    Analyze the following text and return the 3 most relevant main themes or keywords.
    The result must be a valid JSON object with a "themes" key containing a list of strings.
    Ensure the output is *only* the JSON object, with no additional text before or after.

    JSON output example:
    {{
      "themes": ["keyword 1", "keyword 2", "keyword 3"]
    }}

    Text to analyze:
    {content}
    """

    # Preparamos una versi√≥n truncada del contenido para los mensajes de error,
    # para que no imprimamos textos muy largos en la consola.
    truncated_content_for_error = content[:75] + "..." if len(content) > 75 else content


    try:
        api_response = client.messages.create(
            model=ANTHROPIC_MODEL,
            max_tokens=200,
            messages=[
                {"role": "user", "content": prompt_instructions}
            ]
        )

        response_text = api_response.content[0].text.strip()
        json_response = json.loads(response_text)

        if "themes" in json_response and isinstance(json_response["themes"], list):
            extracted_themes = json_response["themes"][:3]
        else:
            print(f"Warning: Unexpected theme format for text: {truncated_content_for_error}. Raw response: {response_text}")
            extracted_themes = [response_text]

    except json.JSONDecodeError as e:
        print(f"Warning: Claude didn't return valid JSON for text: {truncated_content_for_error}. Error: {e}")
        print(f"Raw Claude response: '{response_text}'")
        extracted_themes = ["Error: Invalid JSON from Claude.", response_text]
    except Exception as e:
        print(f"Warning: General error calling Claude API for text: {truncated_content_for_error}. Error: {e}")
        extracted_themes = ["Error extracting themes with Claude."]

    # --- 2.5. Devolver todos los resultados ---
    return {
        "hashtags": hashtags,
        "urls": urls,
        "temas": extracted_themes
    }
```
